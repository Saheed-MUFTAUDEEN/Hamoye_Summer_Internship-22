```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
Energy = pd.read_csv('energydata_complete.csv')
```

```python
Energy 
```

```python
Energy.drop(['date', 'lights'], axis = 1, inplace = True)
```

```python
Energy
```

```python
from sklearn. preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_Energy = pd. DataFrame(scaler.fit_transform(Energy), columns=Energy.columns)
feature_Energy = normalised_Energy.drop(columns=['Appliances'])
variable_target = normalised_Energy['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test, = train_test_split(feature_Energy, variable_target, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
import numpy as np 
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
from sklearn.metrics import r2_score 
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
from sklearn.linear_model import Ridge 
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)
```
